```python
total = df["diagnostico"].value_counts().sum()
print(total)
```

```python
Benignos = df[df['diagnostico'] == 'B']
Malignos = df[df['diagnostico'] == 'M']
total = df["diagnostico"].value_counts().sum()

print(Benignos.shape)
print(Malignos.shape)
print (total)
```

```python
import matplotlib.pyplot as plt

# Contagens
cont_b = Benignos.shape[0]
cont_m = Malignos.shape[0]

# GrÃ¡fico
plt.bar(['Benigno', 'Maligno'], [cont_b, cont_m], color=['green', 'red'])
plt.title('DistribuiÃ§Ã£o de DiagnÃ³sticos')
plt.ylabel('Quantidade')
plt.xlabel('DiagnÃ³stico')
plt.show()
```

```python
df['diagnostico'] = df['diagnostico'].map({'M': 1, 'B': 0})
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='diagnostico', data=df, palette='coolwarm')
plt.title('Contagem de DiagnÃ³stico (0 = Benigno, 1 = Maligno)')
plt.xlabel('DiagnÃ³stico')
plt.ylabel('Contagem')
plt.xticks([0, 1], ['Benigno', 'Maligno'])
plt.show()
```

```python
features_to_plot = ['raio_medio', 'textura_media', 'perimetro_medio', 'area_media', 'concavidade_media']

plt.figure(figsize=(15, 10))
for i, feature in enumerate(features_to_plot):
    plt.subplot(2, 3, i+1)
    sns.histplot(data=df, x=feature, hue='diagnostico', kde=True, palette='coolwarm')
    plt.title(f'DistribuiÃ§Ã£o de {feature}')
    plt.xlabel(feature.replace('_', ' ').title()) # Translate feature name for xlabel
    plt.ylabel('Contagem') # Translate ylabel
    plt.legend(title='DiagnÃ³stico', labels=['Maligno', 'Benigno']) # Translate legend
plt.tight_layout()
plt.show()
```
